```python
df1 = df.iloc[:,0:22]
#create new dataframe that we can mess with and make it clean (we only care about the cols that have data)

df1 = df1.replace('?', np.nan)
df1 = df1.replace('X', np.nan)


print (df1)
#show the new df

NullsPerColumn = df1.isnull().sum()

NullsPerColumn
```

```python



for column in df1:
    binary_encoded = pd.get_dummies(df1.Ethnicity) #we should change '?' to NaNs
    newcols = binary_encoded.columns
    df1[newcols] = binary_encoded
    #make ethnicity into binary dummy variable (better for machine learning later)

for column in df1:
    binary_encoded = pd.get_dummies(df1.Gender)
    newcols = binary_encoded.columns
    df1[newcols] = binary_encoded
    #make gender into binary dummy variable (for ML later)
    
df1.head()
```
